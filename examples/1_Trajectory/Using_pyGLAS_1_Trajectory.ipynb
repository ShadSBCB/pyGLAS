```python
Note the pickled file, GLAS_change_Receptor_GPCRDB.p. pyGLAS generates these files so that you can re-use the results
without having to re-run pyGLAS.
```

```python
Now, as with the single topology example, a single pyGLAS run generates 2 files with the CHICOs and the NACHOs.
```

```python
cat Receptor_GPCRDB_CHICOs.txt
```

```python
It also generates an svg file with the GLAS change in time. Can be viewed using Inkscape (Linux) or Gapplin (Mac).
```

```python
from IPython.display import SVG
SVG("Receptor_GPCRDB_GLAS_overTime.svg")
```

```python
Note that the x-axis is frame number, not time, and that the plot is made up of points, not lines. Lines imply
dependence on previous state; since GLAS is a variable that does not depend on the previous position of the residue
pairs, it is simply plotted as points.
```
